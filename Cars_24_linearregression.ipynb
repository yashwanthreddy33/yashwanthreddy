#steps for multiple linear regression
#1.Before dividing the data frame in to two parts i.e. target and input variables we have to check whether our data frame contains any categorical data or not
#1.if our data frame contains the categorical data then we have to convert the categorical data into continuous data by encoding
#1.when the number of values in the categorical column are limited or only very less then we make use of one hot encoding but if the categorical data ahs large number of values then we make use of target variable encoding
#beacuse your machine learning model is a mathematical model and it only understands the digits not the english words that why we have to convert by using one hot encoding
# target variable encoding: is replacing the categorical column by the mean of target variable.

```python
df["make"] = df.groupby("make")["selling_price"].transform("mean")
```

```python
df["model"] = df.groupby("model")["selling_price"].transform("mean")
```

```python
df
```

```python
# #axis =0 means rows will be drop access =1 means column will be drop
# from sklearn.model_selection import train_test_split
# y=df['selling_price']
# x=df.drop("selling_price",axis=1)
# y.shape,x.shape
# #shape command is used to tell the shape of the dataframe i.e how many rows and how many columns are there
```

```python
#normalization(scaling)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1 = pd.DataFrame(scaler.fit_transform(df),columns = df.columns)
df1
#MinMax scaler is the library which is used for feature scaling
```

```python
#axis =0 means rows will be drop access =1 means column will be drop

y=df1['selling_price']
x=df1.drop("selling_price",axis=1)
y.shape,x.shape
#shape command is used to tell the shape of the dataframe i.e how many rows and how many columns are there
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state =1)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_test_predict = model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
plt.scatter(y_test_predict,y_test)
plt.legend()
plt.show()
```

```python
import seaborn as sns
imp = pd.DataFrame(list(zip(x_test.columns,np.abs(model.coef_))),
                                columns =['Feature','coef'])
sns.barplot(x='Feature',y='coef',data=imp)
plt.xticks(rotation=90)
#zip command is used to pack the different datatypes together
```
