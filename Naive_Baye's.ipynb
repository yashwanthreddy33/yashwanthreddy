```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
```

```python
import pandas as pd
df1 = pd.read_csv('/content/User_Data.csv')
df1.head()
```

```python
df1.shape
```

```python
df1.nunique()
```

```python
df1.isnull().sum()
```

```python
df1.drop(columns =['User ID'],axis=1,inplace=True)
```

```python
df1.columns
```

```python
# Label encoder is the class which is used to convert a categorical variable in to a numerical variable
# Since a machine learning model is a mathematical model so it understands only numerical values.
le = LabelEncoder() # This is also known as one hot encoding
df1['Gender'] = le.fit_transform(df1['Gender']) # There are only two categories i.e.1 and 2.
df1.head()
```

```python
# Split data into dependent/independent variables
x = df1.iloc[:,:-1].values # :-1 Excludes the last column
y = df1.iloc[:,-1].values # -1 only the last column
```

```python
# Split data into test/train set
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=True)
```

```python
# Scale dataset
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
# Classifier
classifier = GaussianNB()
classifier.fit(x_train,y_train)
```

```python
# Prediction
y_pred = classifier.predict(x_test)
# Accuracy
accuracy_score(y_test,y_pred)
```

```python
# Classification report
print(f"Classification Report: \n {classification_report(y_test,y_pred)}")
# support is how many times a particular class is occuring in a dataset
```

```python
# Confusion matrix
cf_matrix = confusion_matrix(y_test,y_pred)
print(cf_matrix)
```
